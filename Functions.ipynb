## Users Defined Functions

```python
#define a function without parameters

def greet_user():
    print("Hello everyone")
```

```python
greet_user()
```

```python
#To define with parameters

def greet_user_personalized(username):
    print(f"Hello {username}")
```

```python
greet_user_personalized("John")
```

```python
def greet_user_personalized():
    username = input("please enter your name: ")
    print(f"Hello {username}")
```

```python
greet_user_personalized()
```

```python
def describe_student(name, age, course):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("Alice", 19, "Python")#positional arguments
```

```python
describe_student(course = "python", name = "Alice", age = 19)#keyword arguments
```

```python
#Defsult values

def describe_student(name, age, course="python"):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("James", 19)
```

```python
describe_student("James", 19, "SQL")
```

# Local vs Global Variables

Any Variable within a fucntion is a local variable

```python
result = 8 #globl result

def sum_two_numbers(a, b):
    result = 5 #local variable
    result = a + b #local variable
    return result
```

```python
print(result)
```

# Passing an arbitrary arguments 

```python
def sum_of_numbers(*numbers):
    result = 0
    for num in numbers:
        result += num
    return result
```

```python
sum_of_numbers(1, 2)
```

```python
sum_of_numbers(1, 2, 10, 5)
```

# Anonymous Functions

```python
#Functions that did not have a name, it's represented by the keyword lambda
```

```python
#To add two variables

add = lambda x, y: x + y
```

```python
print(add(2, 3))
```

```python
print(add(2, 6))
```

```python
power = lambda a, b:a** b
```

```python
print(power(2,3))
```

```python
#Using with filter

numbers = [1, 2, 3, 4, 5]
```

```python
#To filter a list to just even numbers

even_numbers = filter(lambda x: x % 2 == 0, numbers)
```

```python
print(list(even_numbers))
```

```python
odd_numbers = filter(lambda y: y % 2 == 1, numbers)
```

```python
print(list(odd_numbers))
```

```python
#Using with the map function: Map modifies elements in a structure

print(numbers)
```

```python
#To find squares of all the elements in the list

squares = map(lambda x:  x** 2, numbers)
```

```python
print(list(squares))
```
