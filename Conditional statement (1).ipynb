## List Comprehension

```python
squares_comp = [num ** 2 for num in range(1, 11)]


print(squares_comp)

                
```

```python
even_numbers_comp = [num for num in range(0, 21) if num % 2 == 0]

print(even_numbers_comp)
```

## While Loops

To write a code to count from 0 to 5

To write to count from 5 to 1

```python
counter = 1

while counter <= 5:
    print(counter)
    counter += 1
```

To write a code to count from 5 to 1

```python
counter = 5

while counter > 0:
    print(counter)
    counter -= 1
```

break

break is used to exit the loopcompletely, even  the condition is still true.

```python
#To count from 1 to 10, but stop when we reach 5

count = 1

while count <= 10:
    if count == 5:
        break
    print(count)
    count += 1
```

Continue

continue is used to stop the cuurent iteration and move to the next one

```python
#To count 1 to 10 and skip number 5

count = 0

while count < 10:
    count += 1
    if count == 5:
        continue
    print(count)
```
